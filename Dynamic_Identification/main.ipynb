# Exemplo de Otimização

Estimação de parâmetros na identificação de dinâmica de um sistema real (dados foram alterados para manter sigilo, porém dinâmica mantida).

```python
URL_STEP = "https://raw.githubusercontent.com/EduardoPach/Chemical_Engineering/main/Dynamic_Identification/step.csv"
URL_REAL = "https://raw.githubusercontent.com/EduardoPach/Chemical_Engineering/main/Dynamic_Identification/Exemplo_Dados_Reais.csv"
```

```python
data_all = pd.read_csv(URL_REAL)
data_step = pd.read_csv(URL_STEP)
```

```python
data_all['date'] = pd.to_datetime(data_all["date"])
data_step['date'] = pd.to_datetime(data_step["date"])

data_all.set_index("date", inplace=True)
data_step.set_index("date", inplace=True)
```

```python
data_all.plot(subplots=True, figsize=(30,10), title="Dados de toda Operação");
```

```python
data_step.plot(subplots=True, figsize=(30,10), title="Dados na Região do Step");
```

```python
con1 = Control(data=data_step, u="u", y="y1")
con2 = Control(data=data_step, u="u", y="y2")
```

```python
con1.FOPDT_fit(Kp=0, taup=20, thetap=30, plot=True)
```

```python
con2.FOPDT_fit(Kp=0, taup=40, thetap=40, plot=True)
```
